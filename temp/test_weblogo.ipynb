```python
import weblogo
import numpy as np
from io import StringIO
```

```python
pwm_data = """\
A 0.2 0.2 0.3 0.2 0.1
C 0.3 0.2 0.2 0.2 0.4
G 0.3 0.3 0.3 0.2 0.3
T 0.2 0.3 0.2 0.4 0.2
"""

pwm_data
```

```python
fin = StringIO(pwm_data)
fin
```

```python
seqs = weblogo.read_seq_data(fin)
```

```python
seqs
```

```python
options = weblogo.LogoOptions()
options.title = 'PWM Sequence Logo'
options.color_scheme = weblogo.std_color_schemes['classic']
```

```python
logo_data = weblogo.LogoData.from_seqs(seqs)
```

```python
logo_data
```

```python
format = weblogo.LogoFormat(logo_data, options)
png_logo = weblogo.png_formatter(logo_data, format)
```

```python
png_logo
```

```python
# 保存图像
with open('pwm_logo.png', 'wb') as fout:
    fout.write(png_logo)
```
